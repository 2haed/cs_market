```python
cs_knives_lc = [k.lower() for k in cs_knives]

BASE_URL = "https://market.csgo.com/api/full-export/"
EXPORT_FILE = "USD.json"

def fetch_json(url):
    try:
        r = requests.get(url, timeout=10)
        r.raise_for_status()
        return r.json()
    except Exception:
        return None

def is_knife(name):
    return any(knife in name.lower() for knife in cs_knives_lc)

# Загружаем индексный файл
def get_index():
    index_url = BASE_URL + EXPORT_FILE
    index_data = fetch_json(index_url)
    return index_data["format"], index_data["items"]

# Загружаем один item-файл
def parse_item_file(file_name, format_keys):
    url = BASE_URL + file_name
    data = fetch_json(url)
    if not data:
        return []
    filtered = []
    for entry in data:
        item = dict(zip(format_keys, entry))
        # if is_knife(item.get("market_hash_name", "")):
        filtered.append(item)
    return filtered


format_keys, item_files = get_index()
all_items = []

with ThreadPoolExecutor(max_workers=20) as executor:
    futures = [executor.submit(parse_item_file, file_name, format_keys) for file_name in item_files]
    for future in futures:
        result = future.result()
        if result:
            all_items.extend(result)
        time.sleep(0.005)  # небольшая задержка


df = pd.DataFrame(all_items)
df = df[["market_hash_name", "type"]]
df = df[df['type'].isin(['Knife', 'Agent', 'Gloves'])]


```

```python
API_URL = "https://market.csgo.com/api/v2/get-list-items-info"

# Получаем данные по списку hash_name (до 100 штук за раз)
def get_item_info_bulk(hash_names):
    query = "&".join([f"list_hash_name[]={quote(name)}" for name in hash_names])
    url = f"{API_URL}?key={API_KEY}&{query}"
    try:
        r = requests.get(url, timeout=10)
        r.raise_for_status()
        return r.json().get("data", {})
    except Exception:
        return {}

def parse_item_info(data):
    result = {}
    now = time.time()
    week_ago = now - 7 * 24 * 3600
    for name, stats in data.items():
        history = stats.get("history", [])
        average_price = stats.get("average")  # значение от API
        if not history:
            avg_price_7d = None
            sales_7d = 0
            
        else:
            # Фильтруем продажи за последние 7 дней
            recent_sales = [price for ts, price in history if ts >= week_ago]
            sales_7d = len(recent_sales)
            if sales_7d > 0:
                avg_price_7d = round(sum(recent_sales) / sales_7d, 2)
            else:
                avg_price_7d = round(average_price, 2) if average_price else None

        result[name] = {
            "last_price": avg_price_7d,
            "sales_7d": sales_7d,
            "url": f"https://market.csgo.com/en/item/{quote(name)}"
        }
    return result
```

```python
def get_current_prices(currency) -> dict:
    """
    Получаем актуальные цены и объемы продаж по всем предметам.
    Возвращаем словарь: {market_hash_name: {price: float, volume: int}}
    """
    resp = requests.get(f'{BEST_EXPORT_URL + currency}.json')
    data = resp.json()
    if not data.get("success", False):
        return {}

    items = data.get("items", [])
    prices = {}
    for item in items:
        name = item.get("market_hash_name")
        price = float(item.get("price", 0))
        volume = int(item.get("volume", 0))
        prices[name] = {
            "current_price": price,
            "volume": volume
        }
    return prices
```

```python
def get_my_inventory() -> List[Dict]:
    """
    Получаем только те предметы из инвентаря, которые еще не выставлены на продажу.
    """
    url = f"{BASE_URL}/my-inventory?key={API_KEY}"
    try:
        resp = requests.get(url)
        data = resp.json()
        inventory = data.get("items", [])
        current_prices = get_current_prices(currency='RUB')
        for item in inventory:
            name = item.get("market_hash_name")
            price_info = current_prices.get(name)
            if price_info:
                item["current_price"] = price_info["current_price"]
                item["volume"] = price_info["volume"]
            else:
                item["current_price"] = None
                item["volume"] = None
        return inventory
    except Exception:
        return []
```

```python
def add_to_sale(item_id: str, price: float) -> bool:
    """
    Выставляет предмет на продажу. Цена в копейках (1 RUB = 100).
    """
    price_int = int(price * 100)-1  # RUB → копейки
    url = f"{BASE_URL}/add-to-sale"
    params = {
        "key": API_KEY,
        "id": item_id,
        "price": price_int,
        "cur": CURRENCY
    }
    try:
        resp = requests.get(url, params=params)
        return resp.json().get("success", False)
    except Exception:
        return False
```

```python
def mass_add_to_sale(items: list[dict]) -> list[dict]:
    """
    Массово выставляет до 50 предметов на продажу.
    Цена в рублях → копейки (1 RUB = 100), вычитается 1 копейка.
    :param items: список словарей вида {'asset': str/int, 'price': float}
    :return: список результатов для каждого предмета
    """
    if len(items) > 50:
        raise ValueError("Максимум 50 предметов за раз")

    payload = {
        "items": [
            {
                "asset": int(item["asset"]),
                "price": int(item["price"] * 100) - 1  # RUB в копейки минус 1
            }
            for item in items if "asset" in item and "price" in item
        ]
    }

    url = f"{BASE_URL}/mass-add-to-sale"
    params = {
        "key": API_KEY,
        "cur": CURRENCY
    }

    try:
        resp = requests.post(url, params=params, json=payload)
        data = resp.json()
        if not data.get("success", False):
            print("Ошибка при выставлении партии предметов")
            return []
        return data.get("items", [])
    except Exception as e:
        print(f"Ошибка mass_add_to_sale: {e}")
        return []
```

```python
def set_price(item_id: str, price: float, currency: str = CURRENCY) -> bool:
    current_prices = get_current_prices('RUB')
    int_price = int(price * 100)-1  # в копейках
    params = {
        'key': API_KEY,
        'item_id': item_id,
        'price': int_price,
        'cur': currency
    }
    response = requests.get(BASE_URL + 'set-price', params=params)
    data = response.json()
    if data.get("success"):
        print(f"[OK] Установлена цена {price} RUB для item_id={item_id}")
        return True
    else:
        print(f"[ERR] Ошибка для item_id={item_id}: {data.get('error')}")
        return False
```

```python
import requests

def get_usd_to_rub_rate():
    """
    Получает текущий курс доллара от ЦБ РФ и прибавляет 1 рубль.
    """
    url = "https://www.cbr-xml-daily.ru/daily_json.js"
    try:
        r = requests.get(url, timeout=5)
        r.raise_for_status()
        data = r.json()
        usd_rate = data["Valute"]["USD"]["Value"]
        return usd_rate + 1
    except Exception as e:
        print("Ошибка получения курса ЦБ:", e)
        return 95.0  # fallback значение

def convert_last_price_to_usd(all_stats: dict, rub_usd_rate: float) -> dict:
    """
    Переводит last_price из all_stats из копеек в доллары.
    """
    converted_stats = {}
    for name, stat in all_stats.items():
        last_price = stat.get("last_price")
        if last_price:
            usd_price = round(last_price/ rub_usd_rate, 2)
        else:
            usd_price = None
        converted_stats[name] = {
            "url": stat.get("url"),
            "last_price_usd": usd_price,
            "sales_7d": stat.get("sales_7d", 0)
        }
    return converted_stats
from urllib.parse import quote

def generate_url(name):
    encoded_name = quote(name)
    return f"https://market.csgo.com/ru/Knife/{encoded_name}"

```
